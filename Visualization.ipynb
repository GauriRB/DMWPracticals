```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snb
```

```python
d=pd.read_csv('IRIS.csv')
```

# Histogram

```python
d.hist()
```

# Countplot

```python
snb.countplot(data=d,x='petal_length')
```

```python
snb.countplot(data=d,x='petal_length', hue='sepal_length')
```

```python
snb.countplot(data=d,x='species')
plt.title("countplot for species")
```

# Piechart

```python
labels=['setosa','versicolor species','virginica']
sizes=[50,50,50]
```

```python
plt.pie(sizes,labels=labels,explode=(0.1,0.1,0.1))
```

```python
a=np.array(d['petal_length'])
plt.pie(a)
```

# Boxplot

```python
snb.boxplot(data=d,orient='h')
```

# Heatmap

```python
snb.heatmap(d.corr(),cmap='coolwarm',annot=True)
```

# Scatterplot 

```python
snb.jointplot(data=d,x='petal_length',y='sepal_length',kind='scatter')
```

# Barchart

```python
d.plot.bar()
```

# Hexbin

```python
d.plot.hexbin(x='petal_width',y='sepal_width',grid=5)
```
